```python
import binascii
#Manejo de puerto serie
import serial  
import time

############################################
ser = serial.serial_for_url('loop://', timeout=1)

ser.isOpen()
ser.timeout=None
ser.flushInput()
ser.flushOutput()

#code
enable = b'\x80'

opc_1 = b'\x00' 
opc_2 = b'\x01'
opc_3 = b'\x02'
opc_4 = b'\x03'
opc_5 = b'\x04'
opc_6 = b'\x05'
opc_7 = b'\x06'
##############################################

def Codificacion(opc, sub_opc , dispositivo):

    data = opc + enable + b'\x00' + sub_opc

    print(data)

    if (len(data)<=15):
        cabecera = binascii.unhexlify(str(hex(0xA0+eval(hex(len(data))))).split("0x")[1])
        high = b'\x00'
        low  = b'\x00'
        dispo = binascii.unhexlify(dispositivo)
        cola = binascii.unhexlify(str(hex(0x40+eval(hex(len(data))))).split("0x")[1])
        
        Hex_trama = cabecera + high + low + dispo + data + cola

        print("Trama corta en Hex: ", Hex_trama)
        trama_list = list(Hex_trama)

    if (len(data)>15):
        # dispo = opc + enable + b'\x00' + b'\x00' + b'\x00' + b'\x00' + sub_opc
        cabecera = b'\xB0'
        cola = b'\x50'
        high = b'\x01'
        low  = b'\xff' 
        dispo = binascii.unhexlify(dispositivo)

        Hex_trama_large = cabecera + high + low + dispo + data + cola

        print("Trama larga en Hex: ", Hex_trama_large)
        trama_list = list(Hex_trama_large)
        
    return trama_list 

def Decodificacion(trama, opc_deco):

    #CODIFICANDO EL MENSAJE
    mensaje = ser.write(trama)
    print("tamaño:", mensaje)
    print("encode: ", trama) #Mensaje codificado


    large = ser.inWaiting()
    cabecera = 4
    read_data = []
    opc = ''

    if(opc_deco == '1' or '2' or '3' or '4' or '6'):
        while (ser.inWaiting() > 0):  
            read_data.append(chr(ord(ser.read(1))))

        opc = "".join(read_data[cabecera:large-1])
        print("mensaje recibido: ",opc)

    if(opc_deco == '5'):
        for i in range (32000):

            while (ser.inWaiting() > 0):  
                read_data.append(chr(ord(ser.read(1))))

            opc = "".join(read_data[cabecera:large-1])
            print("mensaje recibido: ",opc)

    if(opc_deco == '7'):
        for i in range (8):

            while (ser.inWaiting() > 0):  
                read_data.append(chr(ord(ser.read(1))))

            opc = "".join(read_data[cabecera:large-1])
            print("mensaje recibido: ",opc)

    return opc  

def SubMenu(opc):

    if(opc == '2'):
        
        print("|_____________________parameter options___________________________|")
        print("|( 1 ): Turn on: filter, PRBS, BER, Slizer and counter of Q and I |")
        print("|( 2 ): Trun on: Counter BER Q and I                              |")
        print("|( 3 ): all                                                       |")
        print("|_________________________________________________________________|")
        option = input("Enter an option: ")
        if (option == '1'): return b'\x01'
        if (option == '2'): return b'\x02'
        if (option == '3'): return b'\x03'

    if(opc == '3'):
        
        print("|__Phase options__|")
        print("|( 1 ): Phase (0) |")
        print("|( 2 ): Phase (1) |")
        print("|( 3 ): Phase (2) |")
        print("|( 4 ): Phase (3) |")
        print("|_________________|")
        option = input("Enter an option: ")
        if (option == '1'): return b'\x00'
        if (option == '2'): return b'\x01'
        if (option == '3'): return b'\x02'
        if (option == '4'): return b'\x03'

#main()
decode = ''
dispositivo = '01'
while 1 :
        
    print("|________________MENU PRINCIPAL__________________|")
    print("|( 1 ): Reset globlal                            |")
    print("|( 2 ): Enable filter parameters                 |")
    print("|( 3 ): Phase selection                          |")
    print("|( 4 ): Enable login                             |") #Encenderlo y apagarlo a pata
    print("|( 5 ): Enable read login                        |")
    print("|( 6 ): Bit capture and transmitted errors       |")
    print("|( 7 ): Error or transmitted bits                |")
    print("|( 8 ): Salir del programa                       |")
    print("|________________________________________________|")
    entrada = input("Ingrese algo: ")

    if (entrada  == '8'): ser.close(); break

    if (entrada == '2' or '3' or '7'):  sub_opc = SubMenu(entrada) #Sub menus para opciones

    if(entrada == '1'): trama = Codificacion (opc_1, b'\x00', dispositivo)
    if(entrada == '2'): trama = Codificacion (opc_2, sub_opc, dispositivo)
    if(entrada == '3'): trama = Codificacion (opc_3, sub_opc, dispositivo)
    if(entrada == '4'): trama = Codificacion (opc_4, b'\x01', dispositivo); time.sleep(0.1); Codificacion (opc_4, b'\x00', dispositivo) 
    if(entrada == '5'): trama = Codificacion (opc_5, b'\x01', dispositivo)
    if(entrada == '6'): trama = Codificacion (opc_6, b'\x01', dispositivo)
    if(entrada == '7'): trama = Codificacion (opc_7, b'\x00', dispositivo)

    print("la trama en list es: ", trama)

    decode = Decodificacion(trama, entrada)

```

```python
hex_val1 = 'ab'
hex_val2 = 'ffff'
hex_val3 = 'aaff'

def twosComplement_hex(hexval):
    bits = 16
    val = int(hexval, bits)
    if val & (1 << (bits-1)):
        val -= 1 << bits
    return val
  
print(twosComplement_hex(hex_val1))
print(twosComplement_hex(hex_val2))
print(twosComplement_hex(hex_val3))
```
